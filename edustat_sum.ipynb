```python
edu_sum_seoul = edustat_seoul.groupby(['year', '중계열'], as_index=False).agg(졸업자_남자_비중=('졸업자중남자', 'mean'),
                                  졸업자_여자_비중=('졸업자중여자', 'mean'),
                                  취업자_남자_비중=('취업자중남자', 'mean'),
                                  취업자_여자_비중=('취업자중여자', 'mean'),
                                  전체_취업률=('취업률_계', 'mean'),
                                  남자_취업률=('남자취업률', 'mean'),
                                  여자_취업률=('여자취업률', 'mean'))
```

```python
fig = plt.figure(figsize=(15, 15))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

sns.pointplot(data = edu_sum_seoul, hue = '중계열', x = 'year', y = '졸업자_남자_비중', ax=ax1)
sns.pointplot(data = edu_sum_seoul, hue = '중계열', x = 'year', y = '졸업자_여자_비중', ax=ax1)
sns.pointplot(data = edu_sum_seoul, hue = '중계열', x = 'year', y = '취업자_남자_비중', ax=ax2)
sns.pointplot(data = edu_sum_seoul, hue = '중계열', x = 'year', y = '취업자_여자_비중', ax=ax2)

ax1.set_title('졸업자 중 남녀 비중')
ax2.set_title('취업자 중 남녀 비중')

plt.show()
```
