```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import glob
```

```python
filenames = glob.glob("." + "/i_*.xlsx")
```

```python
df = pd.DataFrame()
dfs = pd.DataFrame()
for f in filenames:
    df1 = pd.read_excel(f, skiprows=1)
    df1['name'] = f
    df = df.append(df1)
    
```

```python
df.columns
```

```python
df.fillna(method = 'ffill', inplace=True)
```

```python
df.to_excel('total_industry.xlsx')
```

```python
total_industry = pd.read_excel('total_industry.xlsx', index_col=0)
```

```python
total_industry
```

```python
from matplotlib import rc
rc('font', family='AppleGothic') 			## 이 두 줄을 
plt.rcParams['axes.unicode_minus'] = False  ## 추가해줍니다. 
```

```python
total_industry = total_industry.assign(남성비율=total_industry.남/total_industry.소계,
                      여성비율=total_industry["학사(여)"] / total_industry.학사)

total_filter = total_industry.rename(columns={'산업별(1)':'산업별'}).\
                              query('산업별 in ["기계", "디스플레이", "반도체", "섬유", "자동차", "전자", "철강", "화학", "바이오ㆍ헬스"]')
```

```python
sns.relplot(data = total_filter, x = '시점', y = '학사(여)', hue = '산업별', col='지역', kind = 'line')
```

```python
sns.relplot(data = total_filter, x = '시점', y = '여성비율', hue = '산업별', col='지역', kind = 'line')
```
